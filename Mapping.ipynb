```python
#workaround for PROJ_LIB error
#you may or may not need this part
import os
import conda

conda_file_dir = conda.__file__
conda_dir = conda_file_dir.split('lib')[0]
proj_lib = os.path.join(os.path.join(conda_dir, 'share'), 'proj')
os.environ["PROJ_LIB"] = proj_lib
```

```python
#standard library imports
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
#declare figure and set size
fig = plt.figure(figsize=(12,6))

#declare map boundaries from corner to corner
#and set resolution
m = Basemap(projection='cyl',
            llcrnrlat = -10,
            urcrnrlat = 60,
            llcrnrlon = -120,
            urcrnrlon = 20,
            resolution = 'l')

#draw coastlines and counties
m.drawcoastlines()
m.drawcountries()

#draw lat and long lines
m.drawparallels(np.arange(-10,60,10),  linewidth=0.1, labels=[True,True,False,False], fontsize=16)
m.drawmeridians(np.arange(-120,20,30), linewidth=0.1, labels=[0,0,1,1], fontsize=16)

# draw a boundary around the map, fill the background.
# this background will end up being the ocean color, since
# the continents will be drawn on top.
m.drawmapboundary(fill_color='aqua')
# fill continents, set lake color same as ocean color.
m.fillcontinents(color='coral',lake_color='aqua')

#plot figure and save to your chosen path
plt.tight_layout()
plt.show()
fig.savefig('/Users/as/Desktop/ColorMap.png')
```
